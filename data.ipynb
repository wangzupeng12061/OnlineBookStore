```python
import pymysql
from faker import Faker
import random

# 初始化Faker实例
fake = Faker()

# 数据库连接配置
db_config = {
    'host': 'localhost',     # 数据库服务器地址
    'user': 'root', # 数据库用户名
    'password': 'wangzupeng1226!', # 数据库密码
    'db': 'book', # 数据库名
    'charset': 'utf8mb4',
    'cursorclass': pymysql.cursors.DictCursor
}

# 生成Suppliers表的数据
def generate_suppliers_data(num_records):
    suppliers_data = []
    for _ in range(num_records):
        name = fake.company()  # 生成供应商名称
        address = fake.address()  # 生成地址
        contact_info = fake.phone_number()  # 生成联系电话

        supplier_record = (name, address, contact_info)
        suppliers_data.append(supplier_record)
    
    return suppliers_data

# 插入数据到数据库
def insert_suppliers_data(db_config, suppliers_data):
    connection = pymysql.connect(**db_config)
    try:
        with connection.cursor() as cursor:
            sql = """
            INSERT INTO Suppliers (Name, Address, ContactInfo) 
            VALUES (%s, %s, %s)
            """
            cursor.executemany(sql, suppliers_data)
        connection.commit()
    finally:
        connection.close()

# 生成并插入数据
num_records = 100  # 可以根据需要调整这个数值
suppliers_records = generate_suppliers_data(num_records)
insert_suppliers_data(db_config, suppliers_records)
print(f"成功插入{num_records}条供应商记录到数据库。")
```

```python
# 书籍和供应商ID范围
book_id_range = (1, 100)  # Books表中有1到100的BookID
supplier_id_range = (1, 100)  # Suppliers表中有1到100的SupplierID

# 生成BookSupplies表的数据
def generate_book_supplies_data(num_records):
    book_supplies_data = []
    generated_combinations = set()

    while len(book_supplies_data) < num_records:
        book_id = random.randint(*book_id_range)
        supplier_id = random.randint(*supplier_id_range)
        
        # 检查组合是否唯一
        if (book_id, supplier_id) not in generated_combinations:
            supply_quantity = random.randint(1, 500)
            supply_date = fake.date_between(start_date='-1y', end_date='today')
            book_supply_record = (book_id, supplier_id, supply_quantity, supply_date)

            book_supplies_data.append(book_supply_record)
            generated_combinations.add((book_id, supplier_id))

    return book_supplies_data

# 插入数据到数据库
def insert_book_supplies_data(db_config, book_supplies_data):
    connection = pymysql.connect(**db_config)
    try:
        with connection.cursor() as cursor:
            sql = """
            INSERT INTO BookSupplies (BookID, SupplierID, SupplyQuantity, SupplyDate) 
            VALUES (%s, %s, %s, %s)
            """
            cursor.executemany(sql, book_supplies_data)
        connection.commit()
    finally:
        connection.close()

# 生成并插入数据
num_records = 1000
book_supplies_records = generate_book_supplies_data(num_records)
insert_book_supplies_data(db_config, book_supplies_records)
print(f"成功插入{num_records}条书籍供应记录到数据库。")
```

```python
# 生成Customers表的数据
def generate_customers_data(num_records):
    customers_data = []
    for _ in range(num_records):
        online_id = fake.user_name()  # 生成在线ID
        password = fake.password()  # 生成密码
        name = fake.name()  # 生成姓名
        address = fake.address()  # 生成地址
        account_balance = round(random.uniform(0, 10000), 2)  # 生成账户余额

        # 根据AccountBalance确定CreditRating
        credit_rating = min(1 + (account_balance // 2000), 5)  # CreditRating最大为5

        customer_record = (online_id, password, name, address, account_balance, credit_rating)
        customers_data.append(customer_record)
    
    return customers_data

# 插入数据到数据库
def insert_customers_data(db_config, customers_data):
    connection = pymysql.connect(**db_config)
    try:
        with connection.cursor() as cursor:
            sql = """
            INSERT INTO Customers (OnlineID, Password, Name, Address, AccountBalance, CreditRating) 
            VALUES (%s, %s, %s, %s, %s, %s)
            """
            cursor.executemany(sql, customers_data)
        connection.commit()
    finally:
        connection.close()

# 生成并插入数据
num_records = 100  # 可以根据需要调整这个数值
customers_records = generate_customers_data(num_records)
insert_customers_data(db_config, customers_records)
print(f"成功插入{num_records}条客户记录到数据库。")
```

```python
# 从Customers表中检索每个客户的ID和地址
def get_customer_addresses(db_config):
    connection = pymysql.connect(**db_config)
    customer_addresses = {}

    try:
        with connection.cursor() as cursor:
            sql = "SELECT CustomerID, Address FROM Customers"
            cursor.execute(sql)
            for row in cursor.fetchall():
                customer_addresses[row['CustomerID']] = row['Address']
    finally:
        connection.close()

    return customer_addresses

# 生成订单
def generate_orders_data(num_records, customer_addresses):
    orders_data = []
    for _ in range(num_records):
        customer_id = random.choice(list(customer_addresses.keys()))
        order_date = fake.date_between(start_date='-1y', end_date='today')
        delivery_address = customer_addresses[customer_id]
        total_amount = round(random.uniform(20, 200), 2)
        delivery_status = random.choice(['Pending', 'Shipped', 'Delivered', 'Cancelled'])

        order_record = (order_date, customer_id, delivery_address, total_amount, delivery_status)
        orders_data.append(order_record)
    
    return orders_data

# 插入数据到数据库
def insert_orders_data(db_config, orders_data):
    connection = pymysql.connect(**db_config)
    try:
        with connection.cursor() as cursor:
            sql = """
            INSERT INTO Orders (OrderDate, CustomerID, DeliveryAddress, TotalAmount, DeliveryStatus) 
            VALUES (%s, %s, %s, %s, %s)
            """
            cursor.executemany(sql, orders_data)
        connection.commit()
    finally:
        connection.close()

# 生成并插入数据
num_records = 1000
customer_addresses = get_customer_addresses(db_config)
orders_records = generate_orders_data(num_records, customer_addresses)
insert_orders_data(db_config, orders_records)
print(f"成功插入{num_records}条订单记录到数据库。")
```

```python
order_id_range = (1, 1000)  # Orders表中有1到1000的OrderID
# 从Books表中获取所有书籍的价格信息
def get_book_prices(db_config):
    connection = pymysql.connect(**db_config)
    book_prices = {}
    
    try:
        with connection.cursor() as cursor:
            sql = "SELECT BookID, Price FROM Books"
            cursor.execute(sql)
            for row in cursor.fetchall():
                book_prices[row['BookID']] = row['Price']
    finally:
        connection.close()

    return book_prices
    
# 生成OrderDetails表的数据
def generate_order_details_data(order_id_range, book_prices):
    order_details_data = []

    for order_id in range(order_id_range[0], order_id_range[1] + 1):
        num_items = random.randint(1, 10)  # 生成1到10个订单详情项
        for _ in range(num_items):
            book_id = random.choice(list(book_prices.keys()))
            quantity = random.randint(1, 5)
            item_amount = round(book_prices[book_id] * quantity, 2)  # 根据书价和数量计算金额

            order_detail_record = (order_id, book_id, quantity, item_amount)
            order_details_data.append(order_detail_record)
    
    return order_details_data

# 插入数据到数据库
def insert_order_details_data(db_config, order_details_data):
    connection = pymysql.connect(**db_config)
    try:
        with connection.cursor() as cursor:
            sql = """
            INSERT INTO OrderDetails (OrderID, BookID, Quantity, ItemAmount) 
            VALUES (%s, %s, %s, %s)
            """
            cursor.executemany(sql, order_details_data)
        connection.commit()
    finally:
        connection.close()

# 获取书籍价格信息
book_prices = get_book_prices(db_config)

# 生成并插入数据
order_details_records = generate_order_details_data(order_id_range, book_prices)
insert_order_details_data(db_config, order_details_records)
print(f"成功插入{len(order_details_records)}条订单详情记录到数据库。")
```

```python
def update_order_total_amounts(db_config):
    connection = pymysql.connect(**db_config)

    try:
        with connection.cursor() as cursor:
            # 获取所有订单ID
            cursor.execute("SELECT OrderID FROM Orders")
            order_ids = [row['OrderID'] for row in cursor.fetchall()]

            for order_id in order_ids:
                # 计算每个订单的总金额
                cursor.execute("""
                    SELECT SUM(ItemAmount) AS TotalAmount
                    FROM OrderDetails
                    WHERE OrderID = %s
                """, (order_id,))
                total_amount = cursor.fetchone()['TotalAmount']

                # 更新Orders表中的total_amount字段
                if total_amount is not None:
                    cursor.execute("""
                        UPDATE Orders
                        SET TotalAmount = %s
                        WHERE OrderID = %s
                    """, (total_amount, order_id))

        connection.commit()
    except Exception as e:
        print(f"更新过程中发生错误: {e}")
    finally:
        connection.close()

# 更新Orders表中的total_amount
update_order_total_amounts(db_config)
```

```python
# 从数据库中获取每本书的订单数量和库存量
def get_book_order_quantities_and_stock(db_config):
    connection = pymysql.connect(**db_config)
    book_order_quantities = {}
    book_stock = {}

    try:
        with connection.cursor() as cursor:
            # 获取每本书的订单总数量
            cursor.execute("""
                SELECT BookID, SUM(Quantity) AS TotalQuantity
                FROM OrderDetails
                GROUP BY BookID
            """)
            for row in cursor.fetchall():
                book_order_quantities[row['BookID']] = row['TotalQuantity']

            # 获取每本书的库存量
            cursor.execute("SELECT BookID, StockQuantity FROM Books")
            for row in cursor.fetchall():
                book_stock[row['BookID']] = row['StockQuantity']

    finally:
        connection.close()

    return book_order_quantities, book_stock
# 生成缺书登记数据    
def generate_book_shortages_data(book_order_quantities, book_stock):
    book_shortages_data = []

    for book_id, stock_quantity in book_stock.items():
        order_quantity = book_order_quantities.get(book_id, 0)
        if order_quantity > stock_quantity:
            required_quantity = order_quantity - stock_quantity
            registration_date = fake.date_between(start_date='-1y', end_date='today')
            processing_status = random.choice(['Pending', 'Ordered', 'Resolved'])

            book_shortage_record = (book_id, registration_date, required_quantity, processing_status)
            book_shortages_data.append(book_shortage_record)
    
    return book_shortages_data
# 插入数据到数据库
def insert_book_shortages_data(db_config, book_shortages_data):
    connection = pymysql.connect(**db_config)
    try:
        with connection.cursor() as cursor:
            sql = """
            INSERT INTO BookShortages (BookID, RegistrationDate, RequiredQuantity, ProcessingStatus) 
            VALUES (%s, %s, %s, %s)
            """
            cursor.executemany(sql, book_shortages_data)
        connection.commit()
    finally:
        connection.close()
# 生成并插入缺书登记数据
book_order_quantities, book_stock = get_book_order_quantities_and_stock(db_config)
book_shortages_records = generate_book_shortages_data(book_order_quantities, book_stock)
insert_book_shortages_data(db_config, book_shortages_records)
print(f"成功插入{len(book_shortages_records)}条缺书登记记录到数据库。")
```

```python
# 获取书籍的供应信息
def get_book_suppliers(db_config):
    connection = pymysql.connect(**db_config)
    book_suppliers = {}

    try:
        with connection.cursor() as cursor:
            sql = "SELECT BookID, SupplierID, SupplyQuantity FROM BookSupplies"
            cursor.execute(sql)
            for row in cursor.fetchall():
                book_id = row['BookID']
                if book_id not in book_suppliers:
                    book_suppliers[book_id] = []
                book_suppliers[book_id].append((row['SupplierID'], row['SupplyQuantity']))
    finally:
        connection.close()

    return book_suppliers

# 获取库存信息
def get_book_stock(db_config):
    connection = pymysql.connect(**db_config)
    book_stock = {}

    try:
        with connection.cursor() as cursor:
            sql = "SELECT BookID, StockQuantity FROM Books"
            cursor.execute(sql)
            for row in cursor.fetchall():
                book_stock[row['BookID']] = row['StockQuantity']
    finally:
        connection.close()

    return book_stock

# 获取缺书信息
def get_book_shortages(db_config):
    connection = pymysql.connect(**db_config)
    book_shortages = {}

    try:
        with connection.cursor() as cursor:
            sql = "SELECT BookID, RequiredQuantity FROM BookShortages"
            cursor.execute(sql)
            for row in cursor.fetchall():
                book_shortages[row['BookID']] = row['RequiredQuantity']
    finally:
        connection.close()

    return book_shortages

# 生成采购订单数据
def generate_purchase_orders_data(book_stock, book_shortages, book_suppliers):
    purchase_orders_data = []

    for book_id in range(*book_id_range):
        quantity = 0
        # 规则1：补充库存至500本
        stock_quantity = book_stock.get(book_id, 0)
        if stock_quantity < 200:
            quantity = 500 - stock_quantity

        # 规则2：补充缺书数量并额外采购300本
        shortage_quantity = book_shortages.get(book_id, 0)
        if shortage_quantity > 0:
            quantity = max(quantity, shortage_quantity + 300)

        if quantity > 0:
            # 选择供应商
            supplier_candidates = book_suppliers.get(book_id, [])
            supplier_id = None
            for supplier, supply_quantity in supplier_candidates:
                if supply_quantity >= quantity:
                    supplier_id = supplier
                    break

            purchase_date = fake.date_between(start_date='-1y', end_date='today')
            arrival_status = random.choice(['Pending', 'Shipped', 'Received'])

            purchase_order_record = (book_id, supplier_id, quantity, purchase_date, arrival_status)
            purchase_orders_data.append(purchase_order_record)
    
    return purchase_orders_data
# 插入数据到数据库
def insert_purchase_orders_data(db_config, purchase_orders_data):
    connection = pymysql.connect(**db_config)
    try:
        with connection.cursor() as cursor:
            sql = """
            INSERT INTO PurchaseOrders (BookID, SupplierID, Quantity, PurchaseDate, ArrivalStatus) 
            VALUES (%s, %s, %s, %s, %s)
            """
            cursor.executemany(sql, purchase_orders_data)
        connection.commit()
    finally:
        connection.close()
# 生成并插入采购订单数据
book_stock = get_book_stock(db_config)
book_shortages = get_book_shortages(db_config)
book_suppliers = get_book_suppliers(db_config)
purchase_orders_records = generate_purchase_orders_data(book_stock, book_shortages, book_suppliers)
insert_purchase_orders_data(db_config, purchase_orders_records)
print(f"成功插入{len(purchase_orders_records)}条采购单记录到数据库。")
```
